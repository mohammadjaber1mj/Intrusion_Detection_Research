```python
# Dropping data (Columns, duplicated rows, NAN, Null..):
from sklearn.utils import shuffle

drop_columns = ["frame.time", "ip.src_host", "ip.dst_host",
                "arp.src.proto_ipv4", "arp.dst.proto_ipv4", "http.file_data",
                "http.request.full_uri", "icmp.transmit_timestamp", "http.request.uri.query",
                "tcp.options", "tcp.payload", "tcp.srcport",
                "tcp.dstport", "udp.port", "mqtt.msg"]

df.drop(drop_columns, axis=1, inplace=True)
df.dropna(axis=0, how='any', inplace=True)
df.drop_duplicates(subset=None, keep="first", inplace=True)
df = shuffle(df)
df.isna().sum()
print(df['Attack_type'].value_counts())
```

```python
from sklearn.preprocessing import StandardScaler, LabelEncoder

encoder = LabelEncoder()

df['http.request.method'] = encoder.fit_transform(df['http.request.method'])
df['http.referer'] = encoder.fit_transform(df['http.referer'])
df['http.request.version'] = encoder.fit_transform(df['http.request.version'])
df['dns.qry.name.len'] = encoder.fit_transform(df['dns.qry.name.len'])
df['mqtt.conack.flags'] = encoder.fit_transform(df['mqtt.conack.flags'])
df['mqtt.protoname'] = encoder.fit_transform(df['mqtt.protoname'])
df['mqtt.topic'] = encoder.fit_transform(df['mqtt.topic'])
df['Attack_type'] = encoder.fit_transform(df['Attack_type'])
```

# Data Poisoning

```python
import numpy as np

def poison_labels(y_train, percentage):
  y_poisoned = y_train.copy()
  n_samples = int(len(y_train) * percentage / 100)

  # Randomly select indices to flip
  indices_to_flip = np.random.choice(y_train.index, size=n_samples, replace=False)

  # Get current labels at these indices
  current_labels = y_train.loc[indices_to_flip].values

  # Prepare all possible labels (once!)
  all_labels = np.array(sorted(y.unique()))

  # Generate new labels ensuring they are different from original labels
  new_labels = np.array([
      np.random.choice(all_labels[all_labels != label])
      for label in current_labels
  ])

  # Assign new labels
  y_poisoned.loc[indices_to_flip] = new_labels

  return y_poisoned
```

```python
poison_percentages = [10, 20, 30, 50]
results_1 = []

for pct in poison_percentages:
  # Poison labels
  y_train_poisoned = poison_labels(y_train, pct)

  # Train XGBoost on poisoned labels

  xgb_model.fit(X_train, y_train_poisoned)

  # Predict on clean test set
  y_pred_poison = xgb_model.predict(X_test)

  results_1.append(y_pred_poison)
```

```python
metrics_results_1 = []

# Loop through predictions and compute metrics
for preds, pct in zip(results_1, poison_percentages):
  acc = accuracy_score(y_test, preds)
  prec = precision_score(y_test, preds, average='macro')
  rec = recall_score(y_test, preds, average='macro')
  f1 = f1_score(y_test, preds, average='macro')

  metrics_results_1.append((pct, acc, prec, rec, f1))
```

```python
# Convert to DataFrame for easy viewing
df_metrics_1 = pd.DataFrame(metrics_results_1, columns=['Percentage', 'Accuracy', 'Precision', 'Recall', 'F1-Score'])
df_metrics_1.head()
```

```python
plt.figure(figsize=(10,6))

plt.plot(df_metrics_1['Percentage'], df_metrics_1['Accuracy'], marker='o', label='Accuracy')
plt.plot(df_metrics_1['Percentage'], df_metrics_1['Precision'], marker='o', label='Precision')
plt.plot(df_metrics_1['Percentage'], df_metrics_1['Recall'], marker='o', label='Recall')
plt.plot(df_metrics_1['Percentage'], df_metrics_1['F1-Score'], marker='o', label='F1-Score')

plt.xlabel('Percentage of Flipped Labels')
plt.ylabel('Score')
plt.title('Performance Degradation under Data Poisoning')
plt.ylim(0.9, 1)
plt.legend()
plt.grid(True)
plt.show()
```

```python
def poison_attack_labels(y, percentage, attack_label=7):
  """
  Poison only attack labels (non-benign) by flipping them to benign class.
  """
  y_poisoned = y.copy()

  # Identify attack samples (non-benign)
  attack_indices = y[y != attack_label].index

  n_samples_to_flip = int(len(attack_indices) * percentage / 100)

  # Randomly select attack indices to flip
  indices_to_flip = np.random.choice(attack_indices, size=n_samples_to_flip, replace=False)

  # Flip attack labels to benign
  y_poisoned.loc[indices_to_flip] = attack_label

  return y_poisoned
```

```python
poison_percentages = [10, 20, 30, 50]
results_2 = []

for pct in poison_percentages:
  # Poison labels
  y_train_poisoned = poison_attack_labels(y_train, pct)

  # Train XGBoost on poisoned labels

  xgb_model.fit(X_train, y_train_poisoned)

  # Predict on clean test set
  y_pred_poison = xgb_model.predict(X_test)

  results_2.append(y_pred_poison)
```

```python
metrics_results_2 = []

# Loop through predictions and compute metrics
for preds, pct in zip(results_2, poison_percentages):
  acc = accuracy_score(y_test, preds)
  prec = precision_score(y_test, preds, average='macro')    # Use macro or weighted as needed
  rec = recall_score(y_test, preds, average='macro')
  f1 = f1_score(y_test, preds, average='macro')

  metrics_results_2.append((pct, acc, prec, rec, f1))
```

```python
def hybrid_poison_labels(y, attack_pct, benign_pct, benign_label=7):
  y_poisoned = y.copy()

  # Flip attack labels to benign
  attack_indices = y[y != benign_label].index
  n_attack_to_flip = int(len(attack_indices) * attack_pct / 100)
  attack_flip_indices = np.random.choice(attack_indices, size=n_attack_to_flip, replace=False)
  y_poisoned.loc[attack_flip_indices] = benign_label

  # Flip benign labels to random attacks
  benign_indices = y[y == benign_label].index
  n_benign_to_flip = int(len(benign_indices) * benign_pct / 100)
  benign_flip_indices = np.random.choice(benign_indices, size=n_benign_to_flip, replace=False)

  # Assign random attack labels to these benign indices
  attack_labels = list(set(y.unique()) - {benign_label})
  y_poisoned.loc[benign_flip_indices] = np.random.choice(attack_labels, size=n_benign_to_flip)

  return y_poisoned
```

```python
poison_percentages = [10, 20, 30, 50]
results_3 = []

for pct in poison_percentages:
  attack_pct = pct/2
  benign_pct = pct/2

  # Poison labels
  y_train_poisoned = hybrid_poison_labels(y_train, attack_pct, benign_pct)

  # Train XGBoost on poisoned labels

  xgb_model.fit(X_train, y_train_poisoned)

  # Predict on clean test set
  y_pred_poison = xgb_model.predict(X_test)

  results_3.append(y_pred_poison)
```

```python
metrics_results_3 = []

# Loop through predictions and compute metrics
for preds, pct in zip(results_3, poison_percentages):
  acc = accuracy_score(y_test, preds)
  prec = precision_score(y_test, preds, average='macro')
  rec = recall_score(y_test, preds, average='macro')
  f1 = f1_score(y_test, preds, average='macro')

  metrics_results_3.append((pct, acc, prec, rec, f1))
```

# Preventing and Detecting Label Manipulation

```python
import pandas as pd
import numpy as np
import hashlib
import rsa
import matplotlib.pyplot as plt
from sklearn.ensemble import IsolationForest

# IIoT Sample Data
sample_df = df.sample(n=200, random_state=42).reset_index(drop=True)
sample_df.head()
```

```python
# Digital Signature Generation and Blockchain Data Recording

(pubkey, privkey) = rsa.newkeys(512)
blockchain_log = []

for idx, row in sample_df.iterrows():
    label = row['Attack_type']
    label_hash = hashlib.sha256(label).hexdigest()
    signature = rsa.sign(label_hash.encode(), privkey, 'SHA-256')

    blockchain_log.append({
        'sample_id': idx,
        'label_hash': label_hash,
        'signature': signature.hex()
    })
```

```python
# Signature Verification

def verify_label(label, entry):
    test_hash = hashlib.sha256(label).hexdigest()
    try:
        rsa.verify(test_hash.encode(), bytes.fromhex(entry['signature']), pubkey)
        return test_hash == entry['label_hash']
    except rsa.VerificationError:
        return False

print("\n Signature Verification Results:")
for entry in blockchain_log:
    label = sample_df.loc[entry['sample_id'], 'Attack_type']
    verified = verify_label(label, entry)
    print(f"Sample {entry['sample_id']}: Verified = {verified}")
```

```python
# Anomaly Detection

# Simulate poisoning: Flip 40% of labels using poison_attack_labels
poisoned_df = sample_df.copy()
poisoned_df['Attack_type'] = poison_attack_labels(poisoned_df['Attack_type'], 40)

# Compare label distribution before and after poisoning
dist_df = pd.DataFrame({
    'Original': sample_df['Attack_type'].value_counts(normalize=True),
    'Poisoned': poisoned_df['Attack_type'].value_counts(normalize=True)
}).fillna(0)

dist_df.plot(kind='bar', title='Label Distribution: Original vs. Poisoned', figsize=(8, 5))
plt.ylabel('Proportion')
plt.xticks(rotation=45)
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
from sklearn.ensemble import IsolationForest
import matplotlib.pyplot as plt

# Copy and encode
encoded_df = sample_df.copy()
encoded_df['class_encoded'] = encoded_df['Attack_type']

# Select only numeric features (excluding label columns)
numeric_cols = encoded_df.select_dtypes(include=[np.number]).columns
feature_cols = numeric_cols.difference(['Attack_type', 'class_encoded'])

# Fit Isolation Forest on features + encoded label
clf = IsolationForest(contamination=0.1, random_state=42)
encoded_df['outlier'] = clf.fit_predict(encoded_df[list(feature_cols) + ['class_encoded']])

# Show outlier results
print("\n Outlier Detection Results (Anomalies flagged as -1):")
print(encoded_df[['Attack_type', 'outlier'] + list(feature_cols)].head())

#Count of outliers
print(f"\nTotal outliers detected: {(encoded_df['outlier'] == -1).sum()} out of {len(encoded_df)}")
```
